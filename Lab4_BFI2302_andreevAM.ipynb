```python
import random
import time
# import copy

def timer(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        value = func(*args, **kwargs)
        print(f"--- {time.time() - start_time:.6f} ms ---")
        return value
    return wrapper

def generation_random_sorted_number(min, max, count):
    data = [random.randint(min, max) for _ in range(count)]
    data.sort()
    print("Items: ", data)
    return data

def Append(data, number):
    data.append(number)
    data.sort()
    print("Item ", number, " added")
    print("Items: ", data)

def Delete(data, number):
    data.remove(number)
    print("Item ", number, " added")
    print("Items: ", data)

class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key
class BinarySearchTree:
    def __init__(self):
        self.root = None
    def insert(self, key):
        if self.root is None:
            self.root = Node(key)
        else:
            self._insert(key, self.root)
    def _insert(self, key, node):
        if key < node.val:
            if node.left is None:
                node.left = Node(key)
            else:
                self._insert(key, node.left)
        elif key > node.val:
            if node.right is None:
                node.right = Node(key)
            else:
                self._insert(key, node.right)
    def search(self, key):
        node = self._search(key, self.root)
        return node.val
    def _search(self, key, node):
        if node is None or node.val == key:
            return node
        elif key < node.val:
            return self._search(key, node.left)
        else:
            return self._search(key, node.right)
    def delete(self, key):
        self.root = self._delete(key, self.root)
    def _delete(self, key, node):
        if node is None:
            return node
        elif key < node.val:
            node.left = self._delete(key, node.left)
        elif key > node.val:
            node.right = self._delete(key, node.right)
        else:
            if node.left is None:
                return node.right
            elif node.right is None:
                return node.left
            else:
                min_node = self._find_min(node.right)
                node.val = min_node.val
                node.right = self._delete(min_node.val, node.right)
        return node
    def find_min(self):
        return self._find_min(self.root)
    def _find_min(self, node):
        if node.left is None:
            return node
        else:
            return self._find_min(node.left)
    def find_max(self):
        return self._find_max(self.root)
    def _find_max(self, node):
        if node.right is None:
            return node
        else:
            return self._find_max(node.right)
```

```python
@timer
def binary_search(arr, num):
    low = 0 
    high = len(arr) - 1 
    while low <= high: 
        mid = (low + high) // 2
        if arr[mid] == num: 
            return mid
        elif arr[mid] > num:
            high = mid - 1
        else:
            low = mid + 1
    return None
@timer
def BST(arr, rand_num):
    bin_tree = BinarySearchTree()
    for num in arr:
        bin_tree.insert(num)
    result = bin_tree.search(rand_num)
    return result
@timer
def fib_search(arr, item):
    fib1, fib2 = 0, 1
    fib_sum = fib1 + fib2
    while fib_sum < len(arr):
        fib1 = fib2
        fib2 = fib_sum
        fib_sum = fib1 + fib2
    offset = -1
    while fib_sum > 1:
        i = min(offset + fib1, len(arr) - 1)
        if arr[i] < item:
            fib_sum = fib2
            fib2 = fib1
            fib1 = fib_sum - fib2
            offset = i
        elif arr[i] > item:
            fib_sum = fib1
            fib2 = fib2 - fib1
            fib1 = fib_sum - fib2
        else:
            return i
    if fib2 == 1 and arr[offset+1] == item:
        return offset + 1
    return None
@timer
def interpolation_search(arr, item):
    low = 0
    high = len(arr) - 1
    while low <= high and item >= arr[low] and item <= arr[high]:
        pos = low + ((item - arr[low]) * (high - low) // (arr[high] - arr[low])) 
        if arr[pos] == item: 
            return pos
        elif arr[pos] < item:
            low = pos + 1
        else: 
            high = pos - 1
    return None

arr = generation_random_sorted_number(1,1000,1000)
rand_num = random.choice(arr)
print("Number chosen: ", rand_num)

start_time = time.time()
builtin = arr.index(rand_num)
print(f"--- {time.time() - start_time:.6f} ms ---")
print('Builtin: ', builtin)
print('Binary search: ', binary_search(arr, rand_num))
print('Binary Tree search: ', BST(arr, rand_num))
print('Fibonacci search: ', fib_search(arr, rand_num))
print('Interpolation: ', interpolation_search(arr, rand_num))
```

```python
class Node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.is_deleted = False

class HashTable_linear:
    def __init__(self):
        self.capacity = 10
        self.size = 0
        self.table = [None] * self.capacity

    def _hash(self, key):
        return hash(key) % self.capacity

    def _rehash(self, old_hash):
        return (old_hash + 1) % self.capacity

    def add(self, key, value):
        if self.size >= 0.7 * self.capacity:
            self._resize()
        hash_idx = self._hash(key)
        start_idx = hash_idx
        while True:
            if self.table[hash_idx] is None or self.table[hash_idx].is_deleted:
                self.table[hash_idx] = Node(key, value)
                self.size += 1
                return
            elif self.table[hash_idx].key == key:
                self.table[hash_idx].value = value
                return
            hash_idx = self._rehash(hash_idx)
            if hash_idx == start_idx:
                self._resize()
                self.add(key, value)
                return
    @timer
    def get(self, key):
        hash_idx = self._hash(key)
        start_idx = hash_idx
        while True:
            node = self.table[hash_idx]
            if node is not None and not node.is_deleted and node.key == key:
                return node.value
            elif node is None:
                return None
            hash_idx = self._rehash(hash_idx)
            if hash_idx == start_idx:
                return None

    def remove(self, key):
        hash_idx = self._hash(key)
        start_idx = hash_idx
        while True:
            node = self.table[hash_idx]
            if node is not None and not node.is_deleted and node.key == key:
                node.is_deleted = True
                self.size -= 1
                return
            elif node is None:
                return
            hash_idx = self._rehash(hash_idx)
            if hash_idx == start_idx:
                return

    def _resize(self):
        old_table = self.table
        old_capacity = self.capacity
        self.capacity *= 2
        self.table = [None] * self.capacity
        self.size = 0
        for node in old_table:
            if node is not None and not node.is_deleted:
                self.add(node.key, node.value)

ht = HashTable_linear()
ht.add("apple", 5)
ht.add("banana", 10)
ht.add("orange", 7)
ht.add("apple", 6)
print('Linear:', ht.get("apple"))
```

```python
class Node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.next = None

class HashTable_chain:
    def __init__(self):
        self.capacity = 10
        self.size = 0
        self.buckets = [None] * self.capacity
  
    def add(self, key, value):
        index = hash(key) % self.capacity
        node = self.buckets[index]
        while node:
            if node.key == key:
                node.value = value
                return
            node = node.next
        new_node = Node(key, value)
        new_node.next = self.buckets[index]
        self.buckets[index] = new_node
        self.size += 1
        if self.size > 0.7 * self.capacity:
            self.resize()
    @timer
    def get(self, key):
        index = hash(key) % self.capacity
        node = self.buckets[index]
        while node:
            if node.key == key:
                return node.value
            node = node.next
        return None
  
    def remove(self, key):
        index = hash(key) % self.capacity
        node = self.buckets[index]
        prev = None
        while node:
            if node.key == key:
                if prev:
                    prev.next = node.next
                else:
                    self.buckets[index] = node.next
                self.size -= 1
                return
            prev = node
            node = node.next
  
    def resize(self):
        new_capacity = self.capacity * 2
        new_buckets = [None] * new_capacity
        for i in range(self.capacity):
            node = self.buckets[i]
            while node:
                index = hash(node.key) % new_capacity
                if new_buckets[index]:
                    new_node = new_buckets[index]
                    while new_node.next:
                        new_node = new_node.next
                    new_node.next = Node(node.key, node.value)
                else:
                    new_buckets[index] = Node(node.key, node.value)
                node = node.next
        self.capacity = new_capacity
        self.buckets = new_buckets
ht = HashTable_chain()
ht.add("apple", 5)
ht.add("banana", 10)
ht.add("orange", 7)
ht.add("apple", 6)
print('Chain:', ht.get("apple"))
```

### Задание №3
Расставить на стандартной 64-клеточной шахматной доске 8 ферзей так, чтобы ни один из них не находился под боем другого». Подразумевается, что ферзь бьёт все клетки, расположенные по вертикалям, горизонталям и обеим диагоналям
Написать программу,  которая находит хотя бы один способ решения задач.

```python
def solve_queens():
    def is_safe(board, row, col):
        for i in range(8):
            if board[row][i] == 1 or board[i][col] == 1:
                return False        
        for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
            if board[i][j] == 1:
                return False
        for i, j in zip(range(row, -1, -1), range(col, 8)):
            if board[i][j] == 1:
                return False
        for i, j in zip(range(row, 8), range(col, -1, -1)):
            if board[i][j] == 1:
                return False
        for i, j in zip(range(row, 8), range(col, 8)):
            if board[i][j] == 1:
                return False
        return True
    def place_queen(board, col):
        if col >= 8:
            return True
        for row in range(8):
            if is_safe(board, row, col):
                board[row][col] = 1
                if place_queen(board, col + 1):
                    return True
                board[row][col] = 0
        return False
    board = [[0 for _ in range(8)] for _ in range(8)]
    if place_queen(board, 0):
        for row in board:
            print(' '.join('Q' if cell else '.' for cell in row))
    else:
        print("No solution")
solve_queens()
```
